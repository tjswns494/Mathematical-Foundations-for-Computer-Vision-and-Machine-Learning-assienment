# import packages for plotting images and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# generate noisy&clean y-coordinate data

```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b
```

# setup matrics

```python
ones=np.ones(num)
A=np.hstack((ones,x))
A=A.reshape(2,num).T
b=np.matrix(y2).T
```

# solve the least square approximate
\^{x} \texttt{=}($A^{T}$A$)^{-1}$${A}^Tb$

```python
lsp=np.linalg.inv(A.T.dot(A)).dot(A.T).dot(b)
```

# find regression line

```python
y=np.array(lsp[0]+lsp[1]*x)
```

# plot the noisy data & clean data

```python
plt.plot(x, y1, 'b.', x, y2, 'y')
plt.show()
```

# plot the line that fits the noisy data by the least square error

```python
plt.plot(x, y1, 'b.', x,y.T,'r')
plt.show()
```
