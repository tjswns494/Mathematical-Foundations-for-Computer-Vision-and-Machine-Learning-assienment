# import packages for plotting images and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define the function

```python
num     = 1001
std     = 5 

def fun(x):
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f
```

# generate noisy&clean y-coordinate data

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

# x  : x-coordinate data
# y1 : (clean) y-coordinate data
# y2 : (noisy) y-coordinate data 
```

# setup matircs

```python
ones=np.ones(num)
b=np.matrix(y2).T

x1= x*x
x2= x1*x
x3=x2*x
x4=x3*x
x5=x4*x
x6=x5*x
x7=x6*x
x8=x7*x
x9=x8*x
```

# solve the least square approximate(p=0)

```python
A=np.column_stack((ones,x))
lsp=np.linalg.inv(A.T.dot(A)).dot(A.T).dot(b)
y=np.array(lsp[0]+lsp[1]*x)
plt.plot(x, y1, 'r',x,y2,'y.', x, y.T,'b')
plt.show()
```

# solve the least square approximate(p=1)

```python
A1=np.column_stack((ones,x,x1))
lsp_1=np.linalg.inv(A1.T.dot(A1)).dot(A1.T).dot(b)
y_1=np.array(lsp_1[0]+lsp_1[1]*x+lsp_1[2]*x1)
plt.plot(x, y1, 'r',x,y2,'y.', x, y_1.T,'b')
plt.show()
```
